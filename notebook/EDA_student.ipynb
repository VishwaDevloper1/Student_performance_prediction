```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('./data/stud.csv')
```

```python
sns.histplot(data=df, x="math_score", hue="gender",kde=True)
```

```python
palette = sns.color_palette("bright",6)
sns.histplot(data=df, x="math_score", hue="parental_level_of_education",kde=True,palette=palette)
```

```python
plt.subplots(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(data=df[df['gender']=='female'], x="math_score", hue="parental_level_of_education",kde=True,palette=palette)
plt.subplot(1,2,2)
sns.histplot(data=df[df['gender']=='male'], x="math_score", hue="parental_level_of_education",kde=True,palette=palette)
```
